```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.path as mpltPath
import math as m
import pandas as pd

from fault_zone_functions import *
```

```python
"""

Adds a low velocity anomaly to the Cascadia Velocity Modle (CVM) to mimic a fault
damage zone. Based on the method presented in Graves & Pitarka, 2016. 
Nominal width of fault zone is 1.5km (750m off either side of fault).

INPUTS:
-CVM. This method is best applied after adding geotech gradients to the CVM.
-Fault file. This contains the xyz locations of fault subsources. Otherwise, a 
    file containing the expected endpoints of the fault should suffice (assuming
    a planar fault trace). Note: if the shallowest subsource is not at the surface,
    the script will extrapolate the surface trace using the specified dip value
-dip value. Average dip value of the fault in the upper 4km of the subsurface. 

OUTPUT:
-CVM with fault zone adjustment

"""
```

```python
#### CHANGE THESE!!! ####

# Input velocity model files
modelTrue = "vs_16l2.nc"
modelTrueVp = "vp_16l2.nc"

# Input fault surface file
faultfile = "./datafiles/faultfiles/tacoma_fault_UTM.txt"
ddip = 45.0 # average fault dip
dstrike = 270.0 # average fault strike

# Fault zone parameters
buff_dist = 750.0 # distance from fault trace to adjust velocity
dist_full = 450.0 # distance from fault trace to apply full velocity reduction
vr = 0.3 # Max value by which to reduce the velocity at the core of the fault zone
vs_min = 200.0 # Do not set velocity lower than this value (assuming value is not -999)

# If your velocity model is very large (i.e., small sampling over a large area),
# processing time can be extremely long. To avoid this, you can create an
# up-sampled grid that corresponds to every 3rd point in your original grid.
up_sample = 0 # 0= no upsample, 1= yes upsample

# Vmodel extent
xmin = 498000 # Tacoma 60X65 model ('TF3' signifier)
xmax = 558000
ymin = 5205200
ymax = 5270000

########################
```

```python
# Open dataset
modelTrue = xr.open_dataset(modelTrue)
modelTrueVp = xr.open_dataset(modelTrueVp)

# Deep copy new object (prevents alteration of original)
model = modelTrue.copy(deep=True)
modelVp = modelTrueVp.copy(deep=True)
```

```python
# Open fault file
# format: [utme umtn z]
ffs = np.loadtxt(faultfile)

# Get top of fault surface
zdep = min(ffs[:,2])
for i in range(len(ffs)):
    if(ffs[i,2]==zdep):
        if(i==0):
            ftrc = ffs[i,0:2]
        else:
            ftrc = np.vstack((ftrc,ffs[i,0:2]))
```

```python
# Find the set of points within 1.5km of the fault trace
# Define polygon around fault trace

nx,ny = len(model.utme.values),len(model.utmn.values)
mod_gridx,mod_gridy = np.meshgrid(model.utme.values,model.utmn.values)
mod_gridx = mod_gridx.flatten()
mod_gridy = mod_gridy.flatten()
mod_grid = np.vstack((mod_gridx,mod_gridy)).T
if(up_sample==1):
    # NOTE: if your utmn and utme arrays are not divisible by 3, the last 1 or 2 rows/columns
    # will not be altered. Assuming your spacing is small relative to the model area, this should
    # not be too much of an issue. 
    newx=np.array([])
    newy=np.array([])
    newix=np.array([])
    newiy=np.array([])
    for i in range(len(model.utme.values)):
        if((i-1)%3==0):
            newx = np.append(newx,model.utme.values[i])
            newix = np.append(newix,i)
    for i in range(len(model.utmn.values)):
        if((i-1)%3==0):
            newy = np.append(newy,model.utmn.values[i])
            newiy = np.append(newiy,i)
    newix=newix.astype(int)
    newiy=newiy.astype(int)
    mod_gridx,mod_gridy = np.meshgrid(newx,newy)
    mod_gridx = mod_gridx.flatten()
    mod_gridy = mod_gridy.flatten()
    mod_grid = np.vstack((mod_gridx,mod_gridy)).T
    nx,ny = len(newx),len(newy)
    
    for i in range(len(model.z)):
        if(model.z.values[i]>=6000.0): # do not apply correction below 6000m
            break
        test_grid = np.zeros((len(newx),len(newy)))
        print("depth = ",model.z.values[i])
        z=model.z.values[i]
        print("Finding points in fault zone")
        poly = buildPolygon(buff_dist,z,zdep,ftrc,dstrike,ddip)
        poly_path = mpltPath.Path(poly)
        inside = np.argwhere(poly_path.contains_points(mod_grid))
        xi = inside - np.floor(inside.astype(np.float32)/float(nx))*nx
        yi = np.floor(inside.astype(np.float32)/float(nx))
        xi = xi.astype(int)
        yi = yi.astype(int)
        dists = np.zeros_like(xi)
        print("# points in fault zone: ",len(xi))
        print("Applying fault zone correction")
        for j in range(len(xi)): # for each point, find distance to outer edge of buffer
            if(j%1000==0):
                print("point "+str(j)+" of "+str(len(xi)))
            distmin = 1000000.0
            for k in range(len(poly)):
                if(k==0): 
                    # Get vector between adjacent polygon vertices (ab)
                    ax,ay = poly[k,0],poly[k,1]
                    bx,by = poly[-1,0],poly[-1,1]
                else:
                    ax,ay = poly[k,0],poly[k,1]
                    bx,by = poly[k-1,0],poly[k-1,1]
                ab = [bx-ax,by-ay]
                mag_ab = m.sqrt(ab[0]**2 + ab[1]**2)

                # Get unit vector perpindicular to ab vector
                if(ab[0]!=0):
                    yy = m.sqrt(1/(ab[1]**2/ab[0]**2+1))
                    xx = -1.0*ab[1]*yy/ab[0]
                else:
                    yy = 0
                    xx = 1
                uv = [xx,yy]

                # Get angle between polygon vertices and point of interest
                ap = [newx[xi[j]]-ax,newy[yi[j]]-ay]
                bp = [newx[xi[j]]-bx,newy[yi[j]]-by]
                thet1 = m.acos(np.dot(uv,ap)/m.sqrt(ap[0]**2 + ap[1]**2))
                thet2 = m.acos(np.dot(uv,bp)/m.sqrt(bp[0]**2 + bp[1]**2))
                # Get distance from vertices to points
                d1 = m.sqrt(ap[0]**2 + ap[1]**2)
                d2 = m.sqrt(bp[0]**2 + bp[1]**2)

                # Get lengths of of triangle legs
                c1 = d1 * m.sin(thet1)
                c2 = d2 * m.sin(thet2)

                # If the condition below is satisfied, 
                # the point lies "between" the vertices.
                # Find the closest distance to the line segment
                # defined by the vertices and update distmin
                if(c1 + c2 - mag_ab < 1e-5): # ran into precision errors when I used 'c1 + c2 == mag_ab', so set this condition instead
                    dis1 = abs(d1*m.cos(thet1))
                    if(dis1<distmin):
                        distmin = dis1
                        dists[j] = dis1
            if(dists[j]>2*buff_dist):
                dists[j] = buff_dist
            if(dists[j]>buff_dist):
                dists[j] = buff_dist - (dists[j] - buff_dist)

            # Apply velocity correction at depth z based on distance from the fault 
            # Apply a full velocity correction at distances less than dist_full from the fault trace
            # Linearly taper to background velocity between dist_full and buff_dist
            if(model.vs.values[newix[xi[j][0]],newiy[yi[j][0]],i]>0): # Only apply correction to points with non-bogus values
                test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2]=1.0
                if(dists[j] > buff_dist - dist_full):
                    test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2] = test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2] - test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2]*vr
                    vs_new = model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] - model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i]*vr
                    vp_new = modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] - modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i]*vr
                else:
                    test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2] = test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2] - test_grid[xi[j][0]-1:xi[j][0]+2,yi[j][0]-1:yi[j][0]+2]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))
                    vs_new = model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] - model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))
                    vp_new = modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] - modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))    

                if(model.z.values[i]>4000 and model.z.values[i]<6000): # Linearly taper to background velocity between 4000m and 6000m
                    vs_new = vs_new * (6000.0 - model.z.values[i])/(6000-4000) + model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] * (1.0 - (6000.0 - model.z.values[i])/(6000-4000))
                    vp_new = vp_new * (6000.0 - model.z.values[i])/(6000-4000) + modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] * (1.0 - (6000.0 - model.z.values[i])/(6000-4000))
                vp_new = np.where(vs_new<vs_min,vs_min*2.5*np.ones_like(vp_new),vp_new)
                vs_new = np.where(vs_new<vs_min,vs_min*np.ones_like(vs_new),vs_new)
                model.vs.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] = vs_new
                modelVp.vp.values[newix[xi[j][0]]-1:newix[xi[j][0]]+2,newiy[yi[j][0]]-1:newiy[yi[j][0]]+2,i] = vp_new
                
        # maintain -999 values if present
        model.vs.values = np.where(modelTrue.vs.values==-999.0,np.ones_like(model.vs.values)*-999.0,model.vs.values)
        modelVp.vp.values = np.where(modelTrueVp.vp.values==-999.0,np.ones_like(modelVp.vp.values)*-999.0,modelVp.vp.values)
                
else:
    for i in range(len(model.z)):
        if(model.z.values[i]>=6000.0): # do not apply correction below 6000m
            break
        test_grid = np.zeros((len(model.utme.values),len(model.utmn.values)))
        print("depth = ",model.z.values[i])
        z=model.z.values[i]
        print("Finding points in fault zone")
        poly = buildPolygon(buff_dist,z,zdep,ftrc,dstrike,ddip)
        poly_path = mpltPath.Path(poly)
        inside = np.argwhere(poly_path.contains_points(mod_grid))
        xi = inside - np.floor(inside.astype(np.float32)/float(nx))*nx
        yi = np.floor(inside.astype(np.float32)/float(nx))
        xi = xi.astype(int)
        yi = yi.astype(int)
        dists = np.zeros_like(xi)
        #print("# polygon edges: ",len(poly))
        print("# points in fault zone: ",len(xi))
        print("Applying fault zone correction")
        for j in range(len(xi)): # for each point, find distance to outer edge of buffer
            if(j%1000==0):
                print("point "+str(j)+" of "+str(len(xi)))
            distmin = 1000000.0
            for k in range(len(poly)):
                if(k==0): 
                    # Get vector between adjacent polygon vertices (ab)
                    ax,ay = poly[k,0],poly[k,1]
                    bx,by = poly[-1,0],poly[-1,1]
                else:
                    ax,ay = poly[k,0],poly[k,1]
                    bx,by = poly[k-1,0],poly[k-1,1]
                ab = [bx-ax,by-ay]
                mag_ab = m.sqrt(ab[0]**2 + ab[1]**2)

                # Get unit vector perpindicular to ab vector
                if(ab[0]!=0):
                    yy = m.sqrt(1/(ab[1]**2/ab[0]**2+1))
                    xx = -1.0*ab[1]*yy/ab[0]
                else:
                    yy = 0
                    xx = 1
                uv = [xx,yy]

                # Get angle between polygon vertices and point of interest
                ap = [model.utme.values[xi[j]]-ax,model.utmn.values[yi[j]]-ay]
                bp = [model.utme.values[xi[j]]-bx,model.utmn.values[yi[j]]-by]
                thet1 = m.acos(np.dot(uv,ap)/m.sqrt(ap[0]**2 + ap[1]**2))
                thet2 = m.acos(np.dot(uv,bp)/m.sqrt(bp[0]**2 + bp[1]**2))
                # Get distance from vertices to points
                d1 = m.sqrt(ap[0]**2 + ap[1]**2)
                d2 = m.sqrt(bp[0]**2 + bp[1]**2)

                # Get lengths of of triangle legs
                c1 = d1 * m.sin(thet1)
                c2 = d2 * m.sin(thet2)

                # If the condition below is satisfied, 
                # the point lies "between" the vertices.
                # Find the closest distance to the line segment
                # defined by the vertices and update distmin
                if(c1 + c2 - mag_ab < 1e-5): # ran into precision errors when I used 'c1 + c2 == mag_ab', so set this condition instead
                    dis1 = abs(d1*m.cos(thet1))
                    if(dis1<distmin):
                        distmin = dis1
                        dists[j] = dis1
            if(dists[j]>2*buff_dist):
                dists[j] = buff_dist
            if(dists[j]>buff_dist):
                dists[j] = buff_dist - (dists[j] - buff_dist)

            # Apply velocity correction at depth z based on distance from the fault 
            # Apply a full velocity correction at distances less than dist_full from the fault trace
            # Linearly taper to background velocity between dist_full and buff_dist
            if(model.vs.values[xi[j],yi[j],i]>0): # Only apply correction to points with non-bogus values
                test_grid[xi[j],yi[j]]=1.0
                if(dists[j] > buff_dist - dist_full):
                    test_grid[xi[j],yi[j]] = test_grid[xi[j],yi[j]] - test_grid[xi[j],yi[j]]*vr
                    vs_new = model.vs.values[xi[j],yi[j],i] - model.vs.values[xi[j],yi[j],i]*vr
                    vp_new = modelVp.vp.values[xi[j],yi[j],i] - modelVp.vp.values[xi[j],yi[j],i]*vr
                else:
                    test_grid[xi[j],yi[j]] = test_grid[xi[j],yi[j]] - test_grid[xi[j],yi[j]]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))
                    vs_new = model.vs.values[xi[j],yi[j],i] - model.vs.values[xi[j],yi[j],i]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))
                    vp_new = modelVp.vp.values[xi[j],yi[j],i] - modelVp.vp.values[xi[j],yi[j],i]*vr*(1.0-(buff_dist-dist_full-dists[j])/(buff_dist-dist_full))    
                if(vs_new<vs_min):
                    vs_new = min(vs_min,model.vs.values[xi[j],yi[j],i])
                    vp_new = 2.5 * vs_new
                if(model.z.values[i]>4000 and model.z.values[i]<6000): # Linearly taper to background velocity between 4000m and 6000m
                    vs_new = vs_new * (6000.0 - model.z.values[i])/(6000-4000) + model.vs.values[xi[j],yi[j],i] * (1.0 - (6000.0 - model.z.values[i])/(6000-4000))
                    vp_new = vp_new * (6000.0 - model.z.values[i])/(6000-4000) + modelVp.vp.values[xi[j],yi[j],i] * (1.0 - (6000.0 - model.z.values[i])/(6000-4000))
                model.vs.values[xi[j],yi[j],i] = vs_new
                modelVp.vp.values[xi[j],yi[j],i] = vp_new
                
        # maintain -999 values if present
        model.vs.values = np.where(modelTrue.vs.values==-999.0,np.ones_like(model.vs.values)*-999.0,model.vs.values)
        modelVp.vp.values = np.where(modelTrueVp.vp.values==-999.0,np.ones_like(modelVp.vp.values)*-999.0,modelVp.vp.values)
```
